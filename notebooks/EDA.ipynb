```python
import os
import random
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from collections import Counter
```

```python
DATASET_PATH="/Users/admin/Desktop/Driver-Behavior-Detection-CNN/data/raw/Revitsone-5classes"
```

```python
print(os.path.exists(DATASET_PATH))
```

```python
classes=sorted(os.listdir(DATASET_PATH))
classes
```

```python
image_counts = {}
for cls in classes:
    folder = os.path.join(DATASET_PATH, cls)
    images = [
        img for img in os.listdir(folder)
        if img.lower().endswith((".jpg", ".jpeg", ".png"))
    ]
    image_counts[cls] = len(images)
image_counts
```

```python
df = pd.DataFrame(
    image_counts.items(),
    columns=["Class","Images"]
)

df
```

```python
print("Total Images :", df["Images"].sum())
```

```python
print("Total Classes :", len(classes))
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["Class"], df["Images"])
plt.title("Class Distribution")
plt.xlabel("Classes")
plt.ylabel("Number of Images")
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig("../outputs/plots/class_distribution.png")
plt.show()
```
